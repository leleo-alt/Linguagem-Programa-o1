```
import pandas as pd
```

```
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

Após executar estas duas células, o DataFrame `df` estará carregado novamente e você poderá prosseguir com suas análises.

```
df.columns = [
    'ano',
    'senioridade',
    'contrato',
    'cargo',
    'salario',
    'moeda',
    'usd',
    'residencia',
    'remoto',
    'empresa',
    'porte_da_empresa'
]
df.head()
```

```
df ["senioridade"].value_counts()
```

```
df ["contrato"].value_counts()
```

```
df ["remoto"].value_counts()
```

```
df ["porte_da_empresa"].value_counts()
```

```
df['senioridade'] = df['senioridade'].map({
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
})
df['senioridade'].value_counts()
```

```
df['contrato'] = df['contrato'].map({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
})
df['contrato'].value_counts()
```
